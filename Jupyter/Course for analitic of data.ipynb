Новый курс - новые знания 
Инструкция + ГК:
a - добавить ячейку выше
b - добавить ячейку ниже
x - вырезать ячейку
v - вставить ячейку
Shift + m - слить ячейки
ctrl + / - закомментировать строку/выделение
ctrl - перескакивать слова. Можно использовать с shift, чтобы выделять
Зажатый alt активирует прямоугольный мультикурсор
Зажатый ctrl активирует мультикурсор
