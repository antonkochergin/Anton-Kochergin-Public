Стек - "очередь" исполняемых комнанд из примера:g ждет print, f ждет g, module ждет f

Напишите реализацию функции closest_mod_5, принимающую в качестве единственного аргумента целое число x и возвращающую самое маленькое целое число y, такое что:

y больше или равно x 

y делится нацело на 5

```python
def closest_mod_5(x):
    for y in range(x,x**2):
        if (y % 5 == 0):
            return y
    return "I don't know :("
x = int(input())
print(closest_mod_5(x))
```

т.к первый аргумент не дописать мы не можем, а второй спокойно можем не вписыать, если зажали ему значение

'*'- дает возможность аргументы, которые не нашли себе место отправить в картеж, список 


```python
def printab(a, b, *args):
    print("positional argument a ", a)
    print("positional argument b ", b)
    print("additional arguments:")
    for arg in args:
        print(arg)

printab(10, 20, 30, 40, 50)
```

то есть в этом случае оставшиеся переменные попадают в слова как ключ + значение 

примечание: если позиционных аргументов больше чем имеется в списке соответвующих этой логике элементов, то не будет работать!

если будет меньше чем заялено, но будет b = 200 в конце например, то функция будет работать исправно

```python
def c(n , k):
    if (k == 0) or (k == n): return 1
    if (k == 1): return n
    if (n < k ): return 0
    if (k>1):
        return c(n-1 , k) + c(n-1,k-1)

arg1,arg2 = map(int , input().split())
print(c(arg1,arg2))
```
